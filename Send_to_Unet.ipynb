```python
from PIL import Image
import os
import numpy as np
```

```python
name_list=[]
for name in os.listdir("/lfs/Nevils/CGAN/art_noise/test_latest/images"):
    if name.endswith("fake_B.png"):
        name_list.append(name)
```

```python
for name in name_list:
    img=np.asarray(Image.open("/lfs/Nevils/CGAN/art_noise/test_latest/images/"+name))[:,:,0]
    img=Image.fromarray(img)
    name=name.strip("_fake_B.png")+".png"
    img.save("/lfs/Nevils/Unet/inputs/art_noise/images/"+name)
```

```python
label_list=[]
for label in os.listdir("/data/Ro_ImageData/32xOTSU"):
    label=label.strip(".tiff")+".png"
    if label in os.listdir("/lfs/Nevils/Unet/inputs/art_noise/images"):
        label_list.append(label)
```

```python
for label in label_list:
    label=label.strip(".png")+".tiff"
    img=Image.open("/data/Ro_ImageData/32xOTSU/"+label)
    img=img.resize((256,256,))
    label=label.strip(".tiff")+".png"
    img.save("/lfs/Nevils/Unet/inputs/art_noise/masks/0/"+label)
```
